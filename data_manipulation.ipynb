```python
import pandas as pd 
import numpy as np
## se define un objeto del tipo dataframe 
df = pd.DataFrame()
## se cargan los datos mediante la función read_excel, al ser un archivo xlsx, mediante el parametro usecols  
## se definen las columnsas a utilizar
df = pd.read_excel('./data/Datos Maestros VF.xlsx',sheet_name = 'Master Data Oficial',
                   usecols=['AGENTE (OFEI)','CENTRAL (dDEC, dSEGDES, dPRU…)',
                            'Tipo de central (Hidro, Termo, Filo, Menor)']) 
## cambiamos el nombre de las columnas para facilitar manipulación 
df=df.set_axis(['Agente','Central','Tipo'],axis='columns')
df
```

```python
## Filtramos los datos segun los requerimientos 
data_fil = df[(df['Agente'] == 'EMGESA S.A.') & ((df['Tipo'] == 'H') | (df['Tipo'] == 'T'))] 
data_fil
```

```python
## Ciclo for para crear una lista con el nombre de las columnas 
columns_names =['Central']
for j in range(1,25):
    columns_names.append('Hora_' + str(j))
print(columns_names)
```

```python
## Cargamos el segundo dataset que es de tipo texto y cambiamos el nombre de las columnas
## mediante el ciclo for implementado arriba 
data_central = pd.DataFrame()
data_central = pd.read_csv('./data/dDEC1204.txt',header = None,encoding='ISO-8859-1')
data_central=data_central.set_axis(columns_names, axis='columns')
data_central
```

```python
## Hacemos merge de los dos dataset 
dmerge=pd.merge(data_fil,data_central)
dmerge
```

```python
## agrupamos los datos por la variable central y sumamos los valores en comun
df_agrupado=dmerge.groupby(['Agente','Central','Tipo']).sum()
df_agrupado
```

```python
## sumamos los valores de manera horizontal y se agrega una columna alfinal con el nombre de total horas
df_agrupado["Total_Horas"] = df_agrupado.sum(axis=1)
df_agrupado
```

```python
## se eliminan aquellos valores iguales a 0 en la columna de total_horas
data_final = df_agrupado[(df_agrupado['Total_Horas'] != 0)]
data_final
```

```python
## exportamos el dataset a un tabla de excel  
writer = data_final.to_excel('./data/consumos_centrales.xlsx',sheet_name='consumos central')
```
